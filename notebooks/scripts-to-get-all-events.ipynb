# Get the Competitions

```python
from statsbombpy import sb
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
comps = sb.competitions()
type(comps)
```

```python
competition_ids = []
season_ids = []
for i in range(len(comps)):
    competition_ids.append(comps.loc[i,"competition_id"])
    season_ids.append(comps.loc[i,"season_id"])
```

# Get Matches

```python
match_ids = []
for c in range(len(competition_ids)):
    comp_id = competition_ids[c]
    seas_id = season_ids[c]
    matches = sb.matches(competition_id=comp_id,season_id=seas_id)
    for m in range(len(matches)):
        match_ids.append(matches.loc[m,"match_id"])
```

# Get Shots

```python
data = []
for e in range(len(match_ids)):
    events = pd.DataFrame(sb.events(match_id=match_ids[e]))
    shots = events[events['type'] == 'Shot']
    shots_df.append(shots)
    data.append(shots)
    if e == 5:
        break

shots_df = pd.concat(data)
```
